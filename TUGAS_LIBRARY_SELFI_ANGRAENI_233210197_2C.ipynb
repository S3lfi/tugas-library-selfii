```python
import matplotlib.pyplot as plt 

# Data contoh 
hari = ['H1', 'H2', 'H3', 'H4', 'H5', 'H6', 'H7', 'H8', 'H9', 'H10']
jumlah_pemasukan = [90, 30, 50, 45, 70, 65, 65, 85, 45, 90]

# grafiknya
plt.figure(figsize=(8, 4))
plt.plot(hari, jumlah_pemasukan, marker= '*', color='purple', linestyle='-')
plt.title('Line Chart Pemasukan Per Hari')
plt.xlabel('hari')
plt.ylabel('Jumlah Pemasukan')
plt.grid(True)
plt.show()
```

```python
Grafik garis di atas merupakan tipe grafik line chart di mana bisa di gunakan untuk melihat jumlah pemasukan per hari.Grafik garis dapat di gunakan
untuk berbagai tujuan di antaranya untuk membandingkan nilai ,seprti nilai pemasukan pada grafik di atas.Secara keseluruhan garis grafik adalah alat untuk
menampilkan data secara visual karna mudah untuk membandingkan nilai.
```

```python
import matplotlib.pyplot as plt


# Data Contoh
bulan = ['B1', 'B2', 'B3', 'B4', 'B5', ' B6', 'B7', 'B8', 'B9', 'B10']
jumlah_penjualan = [10, 30, 40, 60, 70, 80, 40, 90, 10, 80]


#grafiknya
plt.Figure(figsize=(4, 4))
plt.pie(jumlah_penjualan, labels=bulan, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Penjualan per bulan')
plt.axis('equal') # Agar Lingkaran berbentuk sempurna
plt.show()
```

```python
grafik di atas adalah grafik jenis pie chart dimana dapat di gunakan untuk menunjukkan data yang di bagi menjadi beberapa kelompok,seperti data di 
atas menunjukan data hasil dari penjualan perbulan nya sehingga bisa membandingkan nilai dari hasil penjualan.Secara visual grafik tersebut berbentuk
bulat dengan beberapa potongan bagian di mana hasil dari data.
```

```python
import matplotlib.pyplot as plt


# Data Contoh
tahun = ['T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'T7', 'T8', 'T9', 'T10']
jumlah_pengeluaran = [10, 20, 30, 40, 80, 60, 70, 90, 10, 80]


#grafiknya
plt.figure(figsize=(8, 4))
plt.bar(hari, jumlah_penjualan, color='pink')
plt.title('Bar Chart Pengeluaran per tahun')
plt.xlabel('tahun')
plt.ylabel('Jumlah Pengeluaran')
plt.grid(False)
plt.show()
```

```python
Grafik di atas adalah grafik jenis bar chart dimana data yang di bagi menjadi beberapa kategori atau kelompok baris,grafik ini bisa membandingkan kategori
data ,seperti gambar di atas menunjukkan perbandingan data hasil dari pengeluaran pertahunnya,sehingga bisa membandingkan pengeluaran dengan mudah
```
